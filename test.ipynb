```python
def PadMask(padded_input, input_lengths):
    N, T = padded_input.shape[0], padded_input.shape[1]
    # positions: (1, T) vs lengths: (N, 1) -> broadcasts to (N, T)
    positions = torch.arange(T, device=padded_input.device).unsqueeze(0)  # (1, T)
    lengths   = input_lengths.unsqueeze(1)                                  # (N, 1)
    return positions >= lengths  # True where padding
```

## MultiheadAttention

```python
import torch
import torch.nn as nn
```

```python
B = 2 # batch size
L = 5 # query sequence length (target)
S = 5 # key/value sequence length (source)
E = 8 # embeddign dim
H = 2 # heads
```

```python
mha = nn.MultiheadAttention(embed_dim=E, num_heads=H, batch_first=True)
```

```python
query = torch.randn(B,L,E)
key = torch.randn(B,S,E)
value = torch.randn(B,S,E)
```

```python
output, attn_weights = mha(query, key, value)
```
