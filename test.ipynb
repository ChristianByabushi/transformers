```python
def PadMask(padded_input, input_lengths):
    N, T = padded_input.shape[0], padded_input.shape[1]
    # positions: (1, T) vs lengths: (N, 1) -> broadcasts to (N, T)
    positions = torch.arange(T, device=padded_input.device).unsqueeze(0)  # (1, T)
    lengths   = input_lengths.unsqueeze(1)                                  # (N, 1)
    return positions >= lengths  # True where padding
```

```python
import numpy as np

# Parameters from your image
T = 1000
t_target = 100

# Define the linear noise schedule beta_t
beta = np.linspace(0.0001, 0.02, T)

# Calculate alpha and alpha_bar
alpha = 1.0 - beta
alpha_bar = np.cumprod(alpha)

# The variance of p(xt | x0) is (1 - alpha_bar_t)
variance_t100 = 1.0 - alpha_bar[t_target]

print(f"Variance at t=100: {variance_t100:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# Generate samples
np.random.seed(42)
real = np.random.uniform(0, 1, 10000)
fake = np.random.uniform(0, 1, 10000)

# Estimate probability densities
kde_real = gaussian_kde(real)
kde_fake = gaussian_kde(fake)

# Test points
x = np.linspace(0, 1, 200)

# Optimal discriminator
D = kde_real(x) / (kde_real(x) + kde_fake(x))

# Check value at x = 1
x_test = 1.0
D_at_1 = kde_real(x_test) / (kde_real(x_test) + kde_fake(x_test))

print("Discriminator output at x=1:", float(D_at_1))

# Plot
plt.plot(x, D)
plt.axhline(0.5, linestyle="--")
plt.title("Optimal Discriminator when distributions match")
plt.xlabel("x")
plt.ylabel("D*(x)")
plt.show()
```
